```python
import pandas as pd
```

```python
arcr_df = pd.read_csv("arcr.csv")
arcr_df
```

```python
gou_df = pd.read_excel("goutallier.xlsx")
gou_df
```

```python
merged_df = pd.merge(arcr_df, gou_df, on=["일련번호", "수술일자"], how="left")
merged_df
```

```python
# Goutallier 존재하지 않는 행이 많음 (4418)
merged_df["Pre Goutallier (SSP)"].isna().value_counts()
```

```python
merged_df.to_csv("arcr_goutallier.csv", index=False)
```
