```python
# ["일련번호", "수술일자"] 복합키 유일하지 않음
df[["일련번호", "수술일자"]].duplicated().any()
```

```python
# ["일련번호", "수술일자"] 복합키 유일하지 않은 경우 탐색
dup_groups = df[df.duplicated(["일련번호", "수술일자"], keep=False)]
dup_groups
```

```python
conflicts = dup_groups.groupby(["일련번호", "수술일자"]).nunique().reset_index()
conflicts
```

```python
# ["일련번호", "수술일자"] 복합키 중복 행은 동일한 행임을 확인 (conflict row 존재하지 않음)
conflict_rows = conflicts[conflicts.drop(columns=["일련번호", "수술일자"]).gt(1).any(axis=1)]
conflict_rows
```

```python
dedup_df = df.drop_duplicates(subset=["일련번호", "수술일자"])
dedup_df
```

```python
# ["일련번호", "수술일자"] 복합키 유일함 확인
dedup_df[["일련번호", "수술일자"]].duplicated().any()
```

```python
dedup_df.to_csv("arcr.csv", index=False)
```
